```python
#this part is used for the best k features of the data
import numpy as np
import pandas as pd
#Fetching the csv file
carDf = pd.read_csv("/home/tulug/Desktop/CS552HW2/car-pricing-data.csv")
#one hot encoding
one_hot_encoded_data = pd.get_dummies(carDf, columns = ['name', 'fuel','owner','transmission','seller_type'])
one_hot_encoded_data
one_hot_encoded_data.dropna(inplace=True)
one_hot_encoded_data
```

```python
Y = one_hot_encoded_data["selling_price"]
one_hot_encoded_data.drop("selling_price",axis=1,inplace=True)
one_hot_encoded_data
```

```python
#independent variabled
X = one_hot_encoded_data.copy()
X
```

```python
#Removing unit names 
X['mileage'] = X['mileage'].map(lambda x: x.lstrip('+-').rstrip('kmpl'))
X['engine'] = X['engine'].map(lambda x: x.lstrip('+-').rstrip('CC'))
X['max_power'] = X['max_power'].map(lambda x: x.lstrip('+-').rstrip('bhp'))
X['mileage'] = X['mileage'].map(lambda x: x.lstrip('+-').rstrip('km/kg'))
X
```

```python
#converting year to age
X["year"] = 2021 - X["year"]
X
```

```python
X.columns
```

```python
#removing 4933 item from the data frame because of empty string
X = X.drop(labels=4933, axis=0)
```

```python
Y = Y.drop(labels=4933, axis=0)
```

```python
#splitting data as column row
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(X,Y, test_size=0.2, random_state=147)
```

```python
#for extracting best features of given dataset
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_regression
# lets use Select K best features
#k - the number of the important features we aim to find
#f_regression is evaluation function
bestKFeatures=SelectKBest(f_regression,k=5)
fit=bestKFeatures.fit(x_train,y_train)
# find the best features for the given training set
x_best=bestKFeatures.fit_transform(x_train,y_train)
x_best 
```

```python
fit.scores_ # scores for each features

dfscores=pd.DataFrame(fit.scores_)
columns=pd.DataFrame(x_train.columns)
columns
```

```python

featureScoresDF=pd.concat([columns, dfscores], axis=1)
featureScoresDF.columns=["Feature","Score"]
featureScoresDF
```

```python
featureScoresDF.nlargest(5, "Score")
```

```python
#train and test sets are divided according to the selected features
x_train = x_train[['max_power','transmission_Automatic','transmission_Manual','name_BMW X4 M Sport X xDrive20d','engine']]
```

```python
x_test = x_test[['max_power','transmission_Automatic','transmission_Manual','name_BMW X4 M Sport X xDrive20d','engine']]
```

```python
#tranforming the data range between 0 and 1
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler(feature_range=(0,1)) 
scaled_x_train=scaler.fit_transform(x_train)
scaled_x_train
```

```python
#calculation of linear regression
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(scaled_x_train,y_train)
LinearRegression()
```

```python
#calculating of model coefficient
model.coef_
```

```python
#calculation of model intercept
model.intercept_
```

```python
#calculation of the model formula 
_str="y="

for i, m in enumerate(model.coef_):
    _str+="x_{}*{}+".format(i,m)
_str+=str(model.intercept_)
```

```python
print(_str)
```

```python
#transformation of data range between 0 and 1
scaled_x_test=scaler.fit_transform(x_test)
```

```python
#printin the formula
y_predict=model.predict(scaled_x_test)
```

```python
y_predict
```

```python
#converting the y_test data to numpy
df_y_test = y_test.to_numpy()
df_y_test
```

```python
#print the predicted and actual values of the data in the test set

for i in range(len(y_predict)):
    print("Predicted value:", y_predict[i]," Actual value:",y_test.values[i])
```

```python
#mean squared error calculation
MSE = np.square(np.subtract(y_test,y_predict)).mean()
MSE
```

```python
#mean absolute error calculation
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_test, y_predict)
```

```python
#Root-mean-squared error calculation
import math
RMSE = math.sqrt(MSE)
RMSE
```

```python
#f2 score calculation
from sklearn.metrics import r2_score
r2_score(y_test, y_predict)
```

```python
#calculating of difference
difference = np.subtract(df_y_test,y_predict)
difference
```
